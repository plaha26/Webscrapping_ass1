# 2.Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release).

```python
page_imdb=requests.get('https://www.imdb.com/list/ls091520106/')#Sending request to the url and check if we can scrape the data
#get_info=BeautifulSoup(my_url,"html.parser")
page_imdb
```

```python
soup_imdb=BeautifulSoup(page_imdb.content)#getting the content of the webpage
#soup_imdb
```

```python
Top_mov=soup_imdb.find_all("h3",class_="lister-item-header")#Finding the top movies list
#Top_mov
```

```python
year=[]#Storing the year information in alist
for i in Top_mov:
    for j in i.find_all("span"):# The year have span tag, so finding all the span tags
        year.append(j.text.replace("/n","").strip())

Year=[]
for i in range(1,len(year),2):
    Year.append(year[i])
    
len(Year) 
```

```python
Rating=[]#Storing the movie rating information

rate=soup_imdb.find_all("div",class_="ipl-rating-star small")#finding with tag and the class
for i in rate:
    Rating.append(i.text.replace("\n",""))
    
len(Rating)
```

```python
#Creating data frame and printing the data frame
import pandas as pd
Top_Movies=pd.DataFrame({})
Top_Movies["Name"]=movie_name
Top_Movies["Year"]=Year
Top_Movies["Rating"]=Rating
Top_Movies
```

# 4.Write a python program to scrap book name, author name, genre and book review of any 5 books from‘www.bookpage.com’


```python
page=requests.get("https://bookpage.com/reviews")# Get the content of the page
soup=BeautifulSoup(page.content,"html.parser")
```

```python
url_tags=soup.find_all("h4",class_="italic")#Get the urls of reviews
url_tags
```

```python
urls=[]#ger the review links

for i in url_tags:
    for j in i.find_all("a",href=True):
        urls.append(j["href"])
```

```python
urls
```

```python
book_name=[]#Store the book name
for url in urls:
    book=requests.get("https://www.bookpage.com"+url)
    soup=BeautifulSoup(book.content,"html.parser")
    book_name.append(soup.find("h1").text.replace("\n","").replace("★",""))
```

```python
book_name[:5]
```

```python
author_name=[]#Store the author name
for i in urls:
    book=requests.get("https://www.bookpage.com"+i)
    soup=BeautifulSoup(book.content,"html.parser")
    author=soup.find("p", class_="sans bold")
    author_name.append(author.text.replace("\n",""))
```

```python
author_name[:5]
```

```python
review= []#Get the reviews
for i in urls:
    book=requests.get("https://www.bookpage.com"+i)
    soup=BeautifulSoup(book.content,"html.parser")
    rev=soup.find("div",class_="article-body")
    review.append(rev.text.replace("\n",""))
    
    
```

```python
review[:5]
```

```python
import pandas as pd
Book_review=pd.DataFrame({})
Book_review["Book"]=book_name[:5]
Book_review["Author"]=author_name[:5]
Book_review["Review"]=review[:5]
Book_review
```
